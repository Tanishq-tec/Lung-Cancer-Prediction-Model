```python
import pandas as pd
import joblib
```

```python
final_model=joblib.load('lung_cancer.pkl')
print('Model Loaded Successfully')
```

```python
test_input=pd.DataFrame(
    [['M',45,2,1,1,2,1,2,1,2,2,2,1,1,1]],
    columns=['GENDER',
             'AGE',
             'SMOKING',
             'YELLOW_FINGERS',
             'ANXIETY',
             'PEER_PRESSURE',
             'CHRONIC DISEASE',
             'FATIGUE ',
             'ALLERGY ',
             'WHEEZING',
             'ALCOHOL CONSUMING',
             'COUGHING',
             'SHORTNESS OF BREATH',
             'SWALLOWING DIFFICULTY',
             'CHEST PAIN']
)
```

```python
y_pred=final_model.predict_proba(test_input)[0][1]
prob_percent = round(y_pred * 100, 2)

if y_pred < 0.30:
    risk_level = "Minimal Risk"
elif y_pred < 0.60:
    risk_level = "Elevated Risk"
else:
    risk_level = "High Risk"

if y_pred < 0.05:
    diagnosis = "Unlikely to Have Cancer"
elif y_pred < 0.30:
    diagnosis = "Very Low Chance of Cancer"
elif y_pred < 0.60:
    diagnosis = "Moderate Chance of Cancer – Further Checkup Recommended"
else:
    diagnosis = "High Chance of Cancer – Immediate Medical Attention Recommended"

allocation_columns=['LUNG_CANCER']
y_pred_name=pd.DataFrame([[y_pred]],columns=allocation_columns)

print("===== Cancer Prediction Result =====")
print(f"Predicted probability of Cancer: {prob_percent}%")
print(f"Risk level: {risk_level}")
print(f"Diagnosis based on threshold ({CANCER_THRESHOLD*100}%): {diagnosis}")
print("===================================")
```
